```
import pandas as pd
dataset=pd.read_csv('IMDB Dataset.csv')
dataset.head()
```

```
train_set=dataset[:40000]
test_set=dataset[40000:]
```

```
import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```
def dataPreProcess(sentence):
  sentence = re.sub(r'[^\w\s]','', sentence)
  tokens = nltk.word_tokenize(sentence)
  stopWords=set(stopwords.words('english'))
  wordList=[]
  for word in tokens:
    if word not in stopWords:
      wordList.append(word)
  wordList2=[]
  ps=PorterStemmer()
  for word in wordList:
    wordList2.append(ps.stem(word))
  return wordList2
```

```
ds1=dataset[:]
ds1
```

```
ds1['word_list']=ds1['review'].apply(lambda x: dataPreProcess(x))
ds1.head()
```

```
import collections
ds1['frequency_count']=ds1['word_list'].apply(lambda x:collections.Counter(x))
```

```
ds1.head
```

```
ds1['sentiment']=ds1['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
```

```
allFeatures=[]
for i in ds1['word_list']:
  allFeatures+=i
allFeaturesFreq=collections.Counter(allFeatures)
```

```
import operator
allFeaturesFreq=dict(allFeaturesFreq)
allFeaturesFreq=sorted(allFeaturesFreq.items(), key=operator.itemgetter(1),reverse=True)
newFeatures=[]
for i in range(1000):
  newFeatures.append(allFeaturesFreq[i][0])
```

```
df=pd.DataFrame(None,columns=newFeatures)
```

```
for i in range(len(ds1)):
  dd={}
  d=dict(ds1['frequency_count'][i])
  for j in newFeatures:
    try:
      dd[j]=d[j]
    except:
      dd[j]=0
  df=df.append(dd,ignore_index=True)
```
